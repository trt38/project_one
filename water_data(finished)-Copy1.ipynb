## *Setup*

```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import csv
import scipy.stats as stats
from scipy.stats import linregress
```

```python
# Import of Census data and creation of a dataframe

census_data = "../project_one/census_income_data.csv"

census_data_income = pd.read_csv(census_data)
census_data_income.head()
```

```python
# Verification of request success of EPA API data
    
url = "https://enviro.epa.gov/enviro/efservice/SDW_VIOL_ENFORCEMENT/CSV"

response = requests.get(url).json
print(response)
```

```python
# Request of EPA data from API CSV data and conversion to data frame

CSV_URL = "https://enviro.epa.gov/enviro/efservice/SDW_VIOL_ENFORCEMENT/CSV" 

#requesting csv from epa website decoding/reading as csv then converting to a list    
with requests.Session() as s:
    download = s.get(CSV_URL)

    decoded_content = download.content.decode('utf-8')

    cr = csv.reader(decoded_content.splitlines(), delimiter=',')
    my_list = list(cr)
    
# Converting to dataframe    
epa_csv = pd.DataFrame(my_list)
epa_csv.head()
```

## *Data Analysis (Tatiana)*

```python
epa_census_data = epa_census_df
```

```python
# Removal of commas to convert Median Household income into an integer

epa_census_data["Median Household Income"] = epa_census_data["Median Household Income"].str.replace("," , '')

epa_census_data["Median Household Income"] = epa_census_data["Median Household Income"].astype(int)

epa_census_data.head()
```

```python
# Grouping the entire dataframe by state and county

state_county = epa_census_data.groupby(["State", "County"])

print(state_county)

state_county.head()
```

```python
# Counting the number of violations that have occured in each county

viol_counts = epa_census_df.groupby(["State", "County"]).size()
viol_counts
```

```python
# Capturing the median household income by state and county

median_income = state_county["Median Household Income"].median()

median_income.head()
```

```python
# Creating a new DataFrame using both violation count median household income
stcounty_summary_df = pd.DataFrame({"Safety Violations": viol_counts,
                                    "Median Household Income": median_income})
stcounty_summary_df
```

```python
# Reset Index and Create Dataframe

stcounty_reset = stcounty_summary_df.reset_index()

stcounty_reset.head()
```

```python
# Addition of a column that includes an income category to data

bins = [0, 39999, 64999, 89999, 114999, 150000]
group_names = ["< 40K", "40K - 64.9K", "65K - 89.9K", "90K - 114.9K", "115K +"]

stcounty_reset["Income Category"] = pd.cut(stcounty_reset["Median Household Income"], bins, labels=group_names, include_lowest=True)

stcounty_reset.head()
```

```python
# Creation of a scatter plot to assist in visualization of correlation. 

mhi = stcounty_reset["Median Household Income"]
sdwv = stcounty_reset["Safety Violations"]

(slope, intercept, rvalue, pvalue, stderr) = linregress(mhi, sdwv)
regress_values = mhi * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

plt.scatter(mhi, sdwv, marker="o", facecolors="lightblue", alpha=0.75)
plt.plot(mhi,regress_values,"r-")
plt.annotate(line_eq, (100000,200), fontsize=12,color="red")
plt.title("County Median Household Income by # of Safety Violations")
plt.xlabel('Median Household Income')
plt.ylabel('Number of Safety Violations')
print(f"The r-squared is: {rvalue**2}")
plt.show()
```

```python
stcounty_reset.boxplot("Safety Violations", by="Income Category", figsize=(15, 8))
```

```python
# Extract individual groups
group1 = stcounty_reset[stcounty_reset["Income Category"] == "< 40K"]["Safety Violations"]
group2 = stcounty_reset[stcounty_reset["Income Category"] == "40K - 64.9K"]["Safety Violations"]
group3 = stcounty_reset[stcounty_reset["Income Category"] == "65K - 89.9K"]["Safety Violations"]
group4 = stcounty_reset[stcounty_reset["Income Category"] == "90K - 114.9K"]["Safety Violations"]
group5 = stcounty_reset[stcounty_reset["Income Category"] == "115K +"]["Safety Violations"]
```

```python
# Perform the ANOVA
stats.f_oneway(group1, group2, group3, group4, group5)
```

```python
# Creation of dataframe to house total number of safety violations by income category. 

income_cat = stcounty_reset.groupby("Income Category")
income_cat_df = pd.DataFrame(income_cat["Safety Violations"].sum())
income_cat_df = income_cat_df.reset_index()

income_cat_df
```

```python
#Creation of bar chart to display income_cat_df

plt.bar(income_cat_df["Income Category"], income_cat_df["Safety Violations"], color="lightblue", align="center")
plt.title("Total Drinking Safety Violations by Income Category")
plt.xlabel("Income Category")
plt.ylabel("Total Drinking Safety Violations")
```
